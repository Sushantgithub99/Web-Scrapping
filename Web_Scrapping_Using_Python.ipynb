#  Web scrapping using Beautiful Soup


```python
# Importing important libraries
from bs4 import BeautifulSoup
import pandas as pd

## HTML file paths ##(Please upload your directory in colab and provide the path link in html_file_path )
html_file_paths = [
    "/content/Aakarsha Petite Kurtis.html",
    "/content/Akiko PLUS SIZE Cotton Slub Bandhej Print Anarkali Kurta.html",
    "/content/Charvi Drishya Kurtis.html"
]

## An empty list to store extracted data ##
all_data = []


## Iterate over each HTML file path  ##
for html_file_path in html_file_paths:
    ## Read HTML content from file ##
    with open(html_file_path, 'r') as f:
        html_content = f.read()

    ## Parse HTML content ##
    soup = BeautifulSoup(html_content, 'html.parser')

    ## Name & Price of product ##
    listing_name = soup.find('span', class_='Text__StyledText-oo0kvp-0 gpVzIL').text.strip() if soup.find('span', class_='Text__StyledText-oo0kvp-0 gpVzIL') else None
    listing_price = soup.find('div', class_="Card__BaseCard-b3n78k-0 MfhTx ShippingInfo__PriceRow-dflqn4-1 cHzEgM ShippingInfo__PriceRow-dflqn4-1 cHzEgM").text.strip().replace('₹', '').replace(',', '') if soup.find('div', class_="Card__BaseCard-b3n78k-0 MfhTx ShippingInfo__PriceRow-dflqn4-1 cHzEgM ShippingInfo__PriceRow-dflqn4-1 cHzEgM") else None

    ## Extract Url of picture ##
    listing_picture = soup.find('img', class_='lazyload ProductDesktopImage__StyledPerfImage-sc-8sgxcr-1 eexnLF plp-img')
    listing_picture_url = listing_picture['src'] if listing_picture else None

    ## Product details section and extract text ##
    product_details_section = soup.find('div', class_='Card__BaseCard-b3n78k-0 WvMul ProductDescription__DetailsCardStyled-sc-1tmqkzf-0 eHfqpm ProductDescription__DetailsCardStyled-sc-1tmqkzf-0 eHfqpm')
    product_details = product_details_section.text.strip() if product_details_section else None

    ## Seller name & check if supplier is trusted or not ##
    seller_name = soup.find('span', class_="Text__StyledText-oo0kvp-0 fFihtW ShopCardstyled__ShopName-du9pku-6 gdULmH ShopCardstyled__ShopName-du9pku-6 gdULmH").text.strip() if soup.find('span', class_="Text__StyledText-oo0kvp-0 fFihtW ShopCardstyled__ShopName-du9pku-6 gdULmH ShopCardstyled__ShopName-du9pku-6 gdULmH") else None
    trusted_supplier = 1 if soup.find('div', class_='ShopCardstyled__MTrustedBadge-du9pku-10') else 0

    ##  Extract star rating ##
    star_rating = soup.find('span', class_='Text__StyledText-oo0kvp-0 UKsFl').text.strip() if soup.find('span', class_='Text__StyledText-oo0kvp-0 UKsFl') else None

    ## Append extracted data to the list ##
    all_data.append({
        'Listing Name': listing_name,
        'Listing Price': listing_price,
        'Listing Picture URL': listing_picture_url,
        'Product Details': product_details,
        'Seller Name': seller_name,
        'Trusted Supplier': trusted_supplier,
        'Star Rating': star_rating
    })

```

```python
## Create a DataFrame from the list of extracted data ##
df = pd.DataFrame(all_data)

## saving data to csv file ##
df.to_csv('extracted_data.csv', index=False)

## Creating new dataframe df1 to read csv file ##
df1 = pd.read_csv("extracted_data.csv")
df1
```
